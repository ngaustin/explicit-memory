```python
import sys
import os
import numpy as np
from tqdm.notebook import tqdm

os.chdir(os.path.join(sys.path[0], ".."))
os.environ["LOGLEVEL"] = "ERROR"
```

```python
from memory.environment.gym import EpisodicMemoryManageEnv

env = EpisodicMemoryManageEnv(
    capacity={'episodic': 10, 'semantic': 0}, 
    max_history=8, 
    # limits= {"heads":4, "tails":4, "names": 4, "allow_spaces": False}
)

print(env.reset().astype(int))
rewards = 0
for _ in tqdm(range(100)):
    state_numeric, reward, done, info = env.step(0)
    print(state_numeric.astype(int), reward)
    rewards += reward
print(rewards)
print(env.observation_space.shape)
```

```python
from memory.environment.gym import SemanticMemoryManageEnv

env = SemanticMemoryManageEnv(
    capacity={'episodic': 0, 'semantic': 4}, 
    max_history=8, 
    # limits= {"heads":4, "tails":4, "names": 4, "allow_spaces": False}
)
rewards = 0

state_numeric = env.reset().astype(int)
print(state_numeric)

for _ in range(1000):
    action = np.argmin(state_numeric[:, -1])
    state_numeric, reward, done, info = env.step(action)
    print(state_numeric.astype(int), reward)
    rewards += reward
print(rewards)
print(env.observation_space.shape)
```

```python
from memory.environment.gym import SemanticQuestionAnswerEnv

env = SemanticQuestionAnswerEnv(
    capacity={'episodic': 0, 'semantic': 4}, 
    max_history=8, 
    # limits= {"heads":4, "tails":4, "names": 4, "allow_spaces": False}
)
rewards = 0

state_numeric = env.reset().astype(int)
print(state_numeric)

for _ in range(100):
    action = np.argmax(state_numeric[:-1, -1])
    state_numeric, reward, done, info = env.step(action)
    print(state_numeric.astype(int), reward)
    rewards += reward
print(rewards)
print(env.observation_space.shape)
```
